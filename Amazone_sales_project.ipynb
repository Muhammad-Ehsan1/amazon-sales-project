```python
## Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##import DataSet 
df=pd.read_csv(r"F:\Data Analytics\Data Sources\Amazon_Sale_Report.csv")
```

```python
## see no of roww & columns
df.shape
```

```python
##See data of top 5 rows 
print(df.head())
```

```python
##See data of last 5 rows
print(df.tail())
```

```python
##See dataset info
print(df.info())
```

```python
##drop unused column 
df=df.drop(["New","PendingS"],axis=1)
df.info()
```

```python
##See Count of null values in each coulumn
pd.isnull(df).sum()
```

```python
##drop null values 
df=df.dropna()
```

```python
## See Columns in dataset
df.columns
```

```python
## Change data types 
df["ship-postal-code"]=df["ship-postal-code"].astype('int')
df["Date"]=pd.to_datetime(df["Date"])
```

```python
##See dataset info after changing data types
print(df.info())
```

```python
##change column name
df.rename(columns={'Qty' : 'Quantity'}, inplace = True)
```

```python
##See the summary of dataset
df.describe()
```

```python
##creating group_by of size and Quaintity 
S_Qty=df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)

##creating barchart which show relation of quantity and size
plt.figure(figsize=(6,5))
ax=sns.barplot(x='Size',y='Quantity', data=S_Qty,hue ='Size',palette="tab10")

## for giving labels to bars
for containers in ax.containers:
    ax.bar_label(containers)
```

from the above barchart it shows : 
that most people order medium size products

```python
##create a bar chart to show courier status
plt.figure(figsize=(15,5))
ay=sns.countplot(data=df, x='Courier Status',hue= 'Status')

## for giving labels to bars
for bars in ay.containers:
    ay.bar_label(bars)
```

from the above barchart : most of the orders are shipped through the courier

```python
## change datatype of category
df['Category'] = df['Category'].astype(str)

##create histogram of 
plt.figure(figsize=(5, 5))
plt.hist(df['Category'], bins=30, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

from the above graph : 
we see most of the customers buys T-Shirts

```python
## Checking B2B Data  by using pie chart 
B2B_Check = df['B2B'].value_counts()

##  Plot the pie chart
plt.figure(figsize=(4,4))
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
plt.show()
```

from above chart : we can see that maximum i.e. 99.3% of buyers are retailers and 0.8% are B2B buyers

```python
#create scatter plot
plt.figure(figsize=(6,5))
plt.scatter(df['Category'], df['Size'] )
plt.xlabel('Category ')  
plt.ylabel('Size')  
plt.title('Scatter Plot') 
plt.show()
```

```python
#creating top_10_States 
top_10_state = df['ship-state'].value_counts().head(10)

#creating Plot count of cities by state
plt.figure(figsize=(12, 6))
z=sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state',hue ="ship-state")
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```

from the above Graph : you can see that most of the buyers are Maharashtra state

The data analysis reveals that the business has a significant customer base in Maharashtra state, mainly serves retailers, fulfills orders through Amazon, experiences high demand for T-shirts, and sees M-Size as the preferred choice among buyers.
